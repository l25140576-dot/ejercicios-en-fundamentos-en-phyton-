üßÆ Ejercicio 1 ‚Äì Calculadora de factura con IVA

```python
def calcular_factura():

 # INGRESAR DATOS
 producto = input("ingrese nombre del producto:")
 precio = float(input("ingrese el precio unitario del producto:"))
 cantidad = int(input("ingrese el total de productos comprados:"))

#HACER CALCULOS
 subtotal= precio * cantidad
 iva = subtotal * 0.16
 total = subtotal + iva

# IMPRIMIR LOS MENSAJES Y LOS RESULTADOS
 print("|FACTURA|")
 print("  ~Producto:", producto)
 print("  ~Precio unitario:", precio)
 print("  ~Cantidad:", cantidad)
 print("  ~Subtotal:", subtotal)
 print("  ~IVA:", iva)
 print("  ~Total:", total)
calcular_factura()

```

üéü Ejercicio 2 ‚Äì C√°lculo de cableado

```python
def calcular_cableado():
 # PEDIR DATOS PARA LOS CALCULOS

  largo = float(input("INGRESE EL LARGO DE LA HABITACION (m):"))
  ancho = float(input("INGRESE EL ANCHO DE LA HABITACION (m):"))
  cantidad = int(input("INGRESE LA CANTIDAD DE FIBRA INSTALAR (m):"))

# HACER LOS RESPECTIVOS CALCULOS
  perimetro = 2 * (largo + ancho)
  cantidad = perimetro * cantidad


# MOSTRAR RESULTADOS
  print( "     |FIBRA|")
  print("  ~PERIMETRO DE LA HABITACION:", perimetro)
  print("  ~CANTIDAD QUE SE NECESITA:", cantidad)
calcular_cableado()
```

üí∏ Ejercicio 3 ‚Äì Sistema de ventas con Pandas

```python
# IMPORTAR LIBRERIA PANDAS (TABLAS)
import pandas as pd

def sistema_ventas():
  #  MOSTAR EL DF DE UN MENU
  ventas = pd.DataFrame(columns=["Producto", "Precio Unitario", "Cantidad", "Total"])

  # MOSTRAR OPCIONES (menu) agregue los numeros para que sea mas facil para el usuario selecionar
  print("   |MENU|")
  print("1. chilaquiles")
  print("2. refresco")
  print("3. agua")
  print("4. tacos")

  # CREAR UN CICLO PARA LAS OPCIONES
  continuar = True
  while continuar:
    try:
      opcion_input = input("Seleccione una opcion del menu (1-4):")
      if not opcion_input:
          print("Entrada inv√°lida. Por favor, seleccione una opci√≥n.")
          continue
      opcion = int(opcion_input)

      producto = ""
      match opcion:
        case 1:
          producto = "chilaquiles"
        case 2:
          producto = "refresco"
        case 3:
          producto = "agua"
        case 4:
          producto = "tacos"
        case _:
          print("opcion no valida")
          continue # SALTA LA OPCION INVALIDA Y SIGUE

      cantidad = int(input("Ingrese la cantidad de productos:"))
      precio = float(input("Ingrese el precio unitario del producto:"))
      total = cantidad * precio

      venta_nueva = pd.DataFrame([[producto, precio, cantidad, total]],
                                columns=["Producto", "Precio Unitario", "Cantidad", "Total"])
      ventas = pd.concat([ventas, venta_nueva], ignore_index=True)
      print("ventas")

      opcion_continuar = input("¬øQuieres agregar otra venta? (si/no):")
      if opcion_continuar.lower() != "si":
        continuar = False

    except ValueError:
      print("Entrada inv√°lida. Por favor, ingrese un n√∫mero para la opci√≥n o una cantidad/precio v√°lido.")
    except Exception as e:
      print(f"Ocurri√≥ un error: {e}")


  print("       |TOTAL DE VENTAS|")
  print(ventas)
  # VER SI VENTAS ESTA VACIA ANTES DE CALCULAR LA SUMA
  if not ventas.empty:
      print("       |TOTAL DE VENTAS DEL DIA|", ventas["Total"].sum())
  else:
      print("       |TOTAL DE VENTAS DEL DIA| 0")


sistema_ventas()
```

üé¨ Ejercicio 4 ‚Äì Calificaci√≥n de pel√≠culas

```python
def sistema_peliculas():
    peliculas = {
        "1": ["Interestelar", 0],
        "2": ["Avatar", 0],
        "3": ["Matrix", 0],
        "4": ["Titanic", 0],
        "5": ["Inception", 0]
    }

    continuar = True
    while continuar:
        print("\n--- MEN√ö DE PEL√çCULAS ---")
        print("1. Calificar pel√≠cula")
        print("2. Ver recomendaciones")
        print("3. Salir")

        opcion = input("Selecciona una opci√≥n (1-3): ")

        match opcion:
            case "1":
                print("\n--- LISTA DE PEL√çCULAS ---")
                for clave, valor in peliculas.items():
                    print(f"{clave}. {valor[0]} (Calificaci√≥n actual: {valor[1]}‚≠ê)")

                eleccion = input("Selecciona el n√∫mero de la pel√≠cula que deseas calificar: ")

                if eleccion in peliculas:
                    calificacion = int(input("Ingresa una calificaci√≥n (1 a 5): "))
                    peliculas[eleccion][1] = calificacion
                    print(f"Has calificado '{peliculas[eleccion][0]}' con {calificacion} estrellas.")
                else:
                    print("N√∫mero inv√°lido, intenta de nuevo.")

            case "2":
                print("\n--- RECOMENDACIONES ---")
                for valor in peliculas.values():
                    if valor[1] >= 5:
                        print(f"‚≠ê {valor[0]} - {valor[1]} estrellas")
                print("Si no ves ninguna, ¬°califica una con 5 para recomendarla!")

            case "3":
                print("Saliendo del sistema...")
                continuar = False

            case _:
                print("Opci√≥n no v√°lida, intenta otra vez.")

# Llamar la funci√≥n
sistema_peliculas()
```
